```python
#FICO Random Forest and Visualization Project
#For this project I will be exploring publicly available data from LendingClub.com. Lending Club connects people who need money (borrowers) with people who have money (investors). Hopefully, as an investor you would want to invest in people who showed a profile of having a high probability of paying you back. I will try to create a model that will help predict this.

#Lending club had a very interesting year in 2016, so let's check out some of their data and keep the context in mind. This data is from before they even went public.

#I will use lending data from 2007-2010 and be trying to classify and predict whether or not the borrower paid back their loan in full.

#Here are what the columns represent:

#credit.policy: 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.
#purpose: The purpose of the loan (takes values "credit_card", "debt_consolidation", "educational", "major_purchase", "small_business", and "all_other").
#int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
#installment: The monthly installments owed by the borrower if the loan is funded.
#log.annual.inc: The natural log of the self-reported annual income of the borrower.
#dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
#fico: The FICO credit score of the borrower.
#days.with.cr.line: The number of days the borrower has had a credit line.
#revol.bal: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).
#revol.util: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).
#inq.last.6mths: The borrower's number of inquiries by creditors in the last 6 months.
#delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
#pub.rec: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).
```

```python
#Plot of Purpose counts, categorized by color of paid status (not.fully.paid)
#Set figure size for visability of all labels
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=loans,palette='Set1')
```

```python
#Visualization of the trend between FICO score and interest rate

#Joint plot (combination scatter and distribution plot) to analyze trend of FICO score and interest rate
sns.jointplot(x='fico',y='int.rate',data=loans,color='purple')
```

```python
# lmplots to see if the trend differed between not.fully.paid and credit.policy.

#Plot lm plot to visualize trend between paid status (not.fully.paid) and credit policy.
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loans,hue='credit.policy',
           col='not.fully.paid',palette='Set1')
```

```python
#Inspect data again before setting up model
```

```python
loans.info()
```

```python
#Since "Purpose" was categorical, I need to transform this column using dummy variables for sklearn
```

```python
cat_feats = ['purpose'] 
```

```python
#Get dummy variables and create dataset for Random Forest Model
final_data = pd.get_dummies(loans, columns = cat_feats, drop_first = True)
```

```python
final_data.info()
```

```python
#Testing and Training set
```

```python
#Import libraries
from sklearn.model_selection import train_test_split
```

```python
#Create testing and training set
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
#Decision Tree

#To show the difference in a decision tree and random forest, I am going to analyze both
```

```python
#Import libraries
from sklearn.tree import DecisionTreeClassifier
```

```python
#Ceate decision tree classifier
dtree = DecisionTreeClassifier()
```

```python
#View function
dtree.fit(X_train,y_train)
```

```python
#Run decision tree predictions
predictions = dtree.predict(X_test)
```

```python
#load necessary libraies
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#View classification report
print(classification_report(y_test,predictions))
```

```python
#View confusion matrix
print(confusion_matrix(y_test,predictions))
```

```python
#Random Forest

#Final random forest model
```

```python
#load required libraries
from sklearn.ensemble import RandomForestClassifier
```

```python
#Set number of estimators for random forest. I chose 600 for this project
rfc = RandomForestClassifier(n_estimators=600)
```

```python
#View random forest classifier function
rfc.fit(X_train,y_train)
```

```python
#Predict class of paid status for the test data
predictions = rfc.predict(X_test)
```

```python
#Load in necessary libraries
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#View classification report. Classification rate is higher than single decision tree
print(classification_report(y_test,predictions))
```

```python
#View random forest confusion matrix. This is much more accurate than the single decisiont tree
print(confusion_matrix(y_test,predictions))
```
